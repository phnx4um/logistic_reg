## Read the data 1 line at a time

```python
documents = ''
sentiments = ''

with open("./ML - Assignment 2/naive_bayes_data.txt", encoding="utf8") as f:
    for info in f:
        info_splt = info.split(' ', 3)
        document = info_splt[3]
        sentiment = info_splt[1]
        
        documents = documents + document
        sentiments = sentiments + sentiment  + '\n'              
```

```python
print(documents[:2000])
print()
print(sentiments[:40])
```

```python
# preprocess data
# remove punctuations
```

```python
from string import punctuation

# get rid of punctuation
processed_documents = ''.join([c for c in documents if c not in punctuation])
print(processed_documents[:2000])
```

```python
#checking if the number of sentiments and documents are of same lenght
```

```python
list_documents = processed_documents.split('\n')
print(len(list_documents))
list_sentiments = sentiments.split('\n')
print(len(list_sentiments))
```

```python
# just to check the output 
```

```python
print('{} : {} ....'.format(list_sentiments[0], list_documents[0][:100]))
```

```python
#split data int 80:20 ratio
```

```python
train_documents = list_documents[:int(0.8*len(list_documents))] 
train_sentiments = list_sentiments[:int(0.8*len(list_sentiments))]

#to check our model predictions
validate_documents =  list_documents[int(0.8*len(list_documents)):] 
validate_sentiments = list_sentiments[int(0.8*len(list_sentiments)):]

print(len(train_documents))
print(len(train_sentiments))

print(len(validate_documents))
print(len(validate_sentiments))

print()
print('Total: {}'.format(len(train_documents) + len(validate_documents)))
```

```python
from collections import Counter

positive_counts = Counter()
negative_counts = Counter()
total_counts = Counter()
```

```python
for i in range(len(train_documents)):
    words = train_documents[i].split()
    if train_sentiments[i] == 'pos':
        for word in words:
            positive_counts[word] += 1
            total_counts[word] += 1
    else:
        for word in words:
            negative_counts[word] += 1
            total_counts[word] += 1
```

```python
print(positive_counts.most_common())
```

```python
print(negative_counts.most_common())
```
